```python
class FeedForward(Network):
    def __init__(self) -> None:

        self.l1=Linear(3, 4)
        self.l2=Linear(4, 4)
        self.l3=Linear(4, 1)

        Network.__init__(self, vars(locals()['self']))

    def forward(self, x):
        x1=ReLU(self.l1(x))
        x2=ReLU(self.l2(x1))
        x3=Softmax(self.l3(x2))
        return x3
```
